```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
from selenium.webdriver.chrome.service import Service
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
url='https://www.indeed.com/worldwide?from=mobRdr&utm_source=%2Fm%2F&utm_medium=redir&utm_campaign=dt'
driver.get(url)
driver.find_element(By.ID,'onetrust-reject-all-handler').click()
```

we need the list of links that will be used for the driver to scrape the data we need,in this case that list is stored in 'countryLinks' and 'listatrabajos' will be the list of diccionaries where we record the name of the country and the jobs available in each of them


```python
# iterate to get the population and jobs number
countryLinks = driver.find_elements(By.XPATH,'//a[@class="worldwide__link"]')
listatrabajos=[]
```

try and except are needed to block the pop-ups and add other routes because each website varies slightly, this is the best way to avoid a bug in the code.

```python
for link in countryLinks:
        driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
        cname=link.text
        driver.get(str(link.get_attribute('href')))
        time.sleep(2)
        try:
            driver.find_element(By.ID,'onetrust-reject-all-handler').click()
        except:
                pass
        time.sleep(2)
        try:
                driver.find_element(By.XPATH,'//button[@class="icl-CloseButton icl-Card-close"]').click()      #cierra el boton de pop u de google
        except:
                pass
        donde=driver.find_element(By.XPATH,'//div[@class="yosegi-EmbeddedLabelInput yosegi-InlineWhatWhere-where"]')
        donde.click()
        driver.find_element(By.XPATH,'//input[@id="text-input-where"]').send_keys(cname)
        driver.find_element(By.XPATH,'//input[@id="text-input-where"]').send_keys(Keys.ENTER)
        time.sleep(2)
        try:
                try:
                        jobscount=driver.find_element(By.XPATH,'//div[@id="searchCount"]')
                except:
                        jobscount=driver.find_element(By.XPATH,'//div[@class="jobsearch-JobCountAndSortPane-jobCount"]')
        except:
                pass
        try:
                jobscount=jobscount.text
        except:
                pass
        try:
                jobscount=jobscount.split(' ')[-2]
        except:
                pass
        try:
                print(jobscount+' '+cname)
                diccionario={'jobs':jobscount,'pais':cname}
                
                listatrabajos.append(diccionario)
        except:
                pass
# The output below shows the result of the code
```
